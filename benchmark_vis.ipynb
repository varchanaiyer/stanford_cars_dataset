```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os
```

### For Inception model

Inception is a mediocre sized model and hence it takes a while to make a prediction. While the max prediction time is about 76 msecs, the mean is only 66.5 msecs.

While this might seem less, it is actually a fairly long time for autonomous driving based applications. For cases like that, we have other models whose benchmarks we will see after this.

```python
# For Inception

pp_time=np.load(os.path.join('benchmarks', 'inception', 'pp_time.npy'))
p_time=np.load(os.path.join('benchmarks', 'inception', 'p_time.npy'))
t_time=np.load(os.path.join('benchmarks', 'inception', 't_time.npy'))

print('-------------FOR CPU---------')

print('-------------For Inference-------------')
print(f'Total Number of Images : {len(p_time)}')
print(f'Max Inference Time is : {round(np.max(p_time[1:]), 6) * 1000} msecs')
print(f'Min Inference Time is : {round(np.min(p_time[1:]), 6) * 1000} msecs')
print(f'Mean Inference Time is : {round(np.mean(p_time[1:]), 6) * 1000} msecs')

plt.plot(p_time[150:300])
plt.ylabel('Time in secs')
plt.xlabel('Iterations')
plt.show()
```
